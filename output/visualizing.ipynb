Import the necessary modules for the project

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```


**Hypothesis**

Most of the movies that are in the top 1000 IMDB_Rating have English original language.

```python
data = pd.read_csv("drive/MyDrive/Colab Notebooks/csv_data/imdb_top_1000_languges.csv",encoding = "ISO-8859-1")
```

```python
data.head()
```

```python
plt.figure(figsize=(15, 10))
plot = sns.kdeplot(x=data['IMDB_Rating'])
# dropna drops na
plot.axvline(x=data['IMDB_Rating'].mean(), c="red", label="mean")
plot.axvline(x=data['IMDB_Rating'].dropna().median(), c="green", label="median")

plt.legend()
```

```python
plot.figure.savefig("/content/drive/MyDrive/Colab Notebooks/w3_project_pipelines/output/Rating_analysis.svg")
```

```python
plt.figure(figsize=(15, 10))
g = sns.countplot(x=data['Original language'], palette="Set2")
```

```python
g.figure.savefig("/content/drive/MyDrive/Colab Notebooks/w3_project_pipelines/output/Original_language_analysis.svg")
```

```python
plt.figure(figsize=(15, 10))
g = sns.kdeplot(x=data['IMDB_Rating'], hue=data['Original language'])
```

```python
g.figure.savefig("/content/drive/MyDrive/Colab Notebooks/w3_project_pipelines/output/IMDB_Rating_hue_Original_language_analysis.svg")
```
